```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv("../data/raw/filtered.tsv", sep = "\t", index_col = 0)
df["reference"] = df["reference"].apply(lambda x : x.lower())
train_df, val_df = train_test_split(df[["reference", "translation"]], test_size=0.02)
train_df.to_csv("../data/interim/train_df")
val_df.to_csv("../data/interim/val_df")
```
